```python
# simulate polar intensities
x = linspace(0, 2*np.pi, 360)
true = cos(3*x)
pi = random.randn(1000, 360) * 5.0 + true[None,:] + 100.0
```

```python
figure()
scatter(x, pi[0])
plot(x, pi.mean(0), c='r')
show()
```

```python
# add some shot-by-shot "beam jitter"
alpha = 0.05
pi *= (alpha*random.randn(1000) + 1.0)[:,None]
```

```python
def corr_by_shot(pi):
    
    corr = np.zeros(pi.shape[1])
    
    mu = pi.mean(1)
    s2 = pi.std(1) ** 2
    #s2 = ones(pi.shape[0])
    
    for delta in range(pi.shape[1]):
        corr[delta] = mean( ((pi-mu[:,None]) * (roll(pi, delta, axis=1)-mu[:,None])) / s2[:,None] )
    
    return corr


def corr_ensemb(pi):
    
    corr = np.zeros(pi.shape[1])
    
    mu = pi.mean(1)
    s2 = pi.std() ** 2
    #s2 = 1.0
    
    print mu, s2
    
    for delta in range(pi.shape[1]):
        corr[delta] = mean( ((pi-mu[:,None]) * (roll(pi, delta, axis=1)-mu[:,None])) / s2 )

    return corr
```

```python
tc = zeros(pi.shape[1])
for d in range(pi.shape[1]):
    tc[d] = correlate(true, np.roll(true, d))
tc /= tc[0]

cbs = corr_by_shot(pi)
cbe = corr_ensemb(pi)
```

```python
figure()
plot(x, cbs, lw=2)
plot(x, cbe, lw=2)
#plot(x, tc, lw=2)
#ylim([-.1,.1])
legend(['shot-by-shot norm', 'ensemble norm', 'true correlation'])
show()
```
