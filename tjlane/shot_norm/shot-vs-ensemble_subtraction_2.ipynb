```python
# generate a "true" signal

n_pix = 360
x = linspace(0, 2*pi, n_pix)
c = cos(3*x)

figure()
plot(x, c, lw=2)
show()
```

```python
# generate 100 instances w/gaussian noise
obs = c[None,:] + random.randn(250, n_pix) * 15.0


figure()
scatter(x, obs[0])
plot(x, obs.mean(0), lw=2, color='r')
show()
```

```python
# define correlation fxns & error function

# shot-by-shot

def correlate(x, mask, subtraction='shot'):
    """
    Autocorrelate x

    subtraction = {'shot', 'ensemble'}
    """
    
    x2 = x.copy()
    N = x.shape[1]
    
    if subtraction == 'shot':
        x_bar = x2[:,:] - x2[:,:].mean(1)[:,None]
    elif subtraction == 'ensemble':
        x_bar = x2[:,:] - mean( x2[:,:] )

    corr = np.zeros((x.shape[0], N))
    for k in range(x.shape[0]):
        for delta in range(N):
            for i in range(N):
        
                i_shifted = (i+delta) % N
                if mask[i] and mask[i_shifted]:
                    corr[k, delta] += x_bar[k,i] * x_bar[k,i_shifted] # avg over shots
            
    return corr


def rmse(signal, reference):
    assert signal.shape[1] == len(reference)
    N = float(len(reference))
    return np.sqrt( (1.0 / N) * np.sum( np.power(signal-reference[None,:], 2) , axis=1 ) )
```

```python
# plot the spread in "shot means"
shot_means = obs[:,:].mean(1)
ensb_mean  = mean( obs[:,:] )

figure()
hist(shot_means, 25, histtype='step', lw=2)
vlines(ensb_mean, 0, 10.0, color='r', lw=2)
show()
```

```python
# perform both kinds of correlations

mask = np.ones(n_pix)

shot_by_shot = correlate(obs, mask, subtraction='shot')
ensemble     = correlate(obs, mask, subtraction='ensemble')

assert np.all(shot_by_shot != ensemble)
print shot_by_shot.shape
```

```python
# measure the errors
shot_error = rmse(shot_by_shot, c)
ensb_error = rmse(ensemble, c)

figure()

hist(shot_error, 25, histtype='step', lw=2)
hist(ensb_error, 25, histtype='step', lw=2)
legend(['shot subtraction', 'ensemble subtraction'])

xlabel('RMSE to converged correlator')
ylabel('Shot Correlator Frequency')

ylim([0,30])

savefig('sinewave_shot_vs_ensemble_subtraction.pdf')
show()

print "Shot-mean expected error: %f / std: %f" % (shot_error.mean(), shot_error.std())
print "Ensb-mean expected error: %f / std: %f" % (ensb_error.mean(), ensb_error.std())
```
