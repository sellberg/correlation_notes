##Simulations to test different statistical estimators of CXS correlation functions

I set out to determine which correlation function estimator gives the best signal. There has been some argument in the group about this -- these simulations will hopefully help clear this up.

There are 3 arguments:
   <br>(1) whether to compute fluctuations around the "shot" mean or population/"ensemble" mean
   <br>(2) whether to normalize on a shot-by-shot or ensemble basis
   <br>(3) whether to normalize using std deviations, means, or sqrt means
    
See TJL's report for a more thorough discussion of what these mean and the differences

```python
from odin import xray
from odin import structure

xray.logger.propagate = False
xray.scatter.logger.propagate = False

gold = structure.load_coor('fcc_sphere_5.0nm.coor')
```

```python
# simulate a 5nm gold NP at q=2.67 inv ang into a polar 360 px detector. 1 molecule/shot, 1k shots
r = xray.Rings.simulate(gold, 1, [2.67], 360, 1000)
```

```python
# add a random mask -- each px has a 10% chance of being "bad"
polar_mask = random.binomial(1, 0.9, size=(1, 360))
```

```python
# code for the SUBTRACTION comparison

# compute correlators, both subtracting shot-by-shot means
# and total means

# will do this for the first ring (q=2.67) only

# shot-by-shot
x_bar = r.polar_intensities[:,0,:] - r.polar_intensities[:,0,:].mean(1)[:,None]

sbs_corr = np.zeros(r.num_phi)
for delta in range(r.num_phi):
    for i in range(r.num_phi):
        
        i_shifted = (i+delta) % r.num_phi
        if polar_mask[0,i] and polar_mask[0,i_shifted]:
            sbs_corr[delta] += mean( x_bar[:,i] * x_bar[:,i_shifted] ) # avg over shots
            
# total
x_bar = r.polar_intensities[:,0,:] - mean( r.polar_intensities[:,0,:] )

ttl_corr = np.zeros(r.num_phi)
for delta in range(r.num_phi):
    for i in range(r.num_phi):
        
        i_shifted = (i+delta) % r.num_phi
        if polar_mask[0,i] and polar_mask[0,i_shifted]:
            ttl_corr[delta] += mean( x_bar[:,i] * x_bar[:,i_shifted] ) # avg over shots

```

```python
# plot the resulting correlation functions against the correlation w/o a mask

# normalization schemes
sbs_mean_corr = r.correlate_intra(2.67, 2.67, 0, mean_only=True)
sbs_std_corr = r.correlate_intra(2.67, 2.67, 1, mean_only=True)
ens_mean_corr = r.correlate_intra(2.67, 2.67, 2, mean_only=True)
ens_std_corr = r.correlate_intra(2.67, 2.67, 3, mean_only=True)

# subtraction schemes
sbs_true_corr = r.correlate_intra(2.67, 2.67, 4, mean_only=True)

x_bar = r.polar_intensities[:,0,:] - mean( r.polar_intensities[:,0,:] )
ttl_true_corr = np.zeros(r.num_phi)
for delta in range(r.num_phi):
    for i in range(r.num_phi):
        i_shifted = (i+delta) % r.num_phi
        ttl_true_corr[delta] += mean( x_bar[:,i] * x_bar[:,i_shifted] ) # avg over shots
```

```python
figure(figsize=(12,5))

subplot(121)
title('Shot-by-Shot Subtraction')
plot(sbs_corr / sbs_corr[0], lw=2)
plot(sbs_true_corr / sbs_true_corr[0], lw=2)
legend(['with gaps', 'no gaps'])
xlim([0, r.num_phi])
ylim([-0.05, 0.2])

subplot(122)
title('Ensemble Mean Subtraction')
plot(ttl_corr / ttl_corr[0], lw=2)
plot(ttl_true_corr / ttl_true_corr[0], lw=2)
legend(['with gaps', 'no gaps'])
xlim([0, r.num_phi])
ylim([-0.05, 0.2])

savefig('comp_subtraction_gaps.pdf')
show()
```

```python
# plot different normalization schemes
figure()

title('Comparison of Normalization Schemes')

plot(sbs_mean_corr / sbs_mean_corr[0], lw=2)
plot(sbs_std_corr / sbs_std_corr[0], lw=2)
plot(ens_mean_corr / ens_mean_corr[0], color='teal', lw=2)
plot(ens_std_corr / ens_std_corr[0], '--', color='purple', lw=2)

legend(['shot-by-shot mean (0)', 'shot-by-shot std (1)', 'ensb mean (2)', 'ensb std (3)'])

xlim([0, r.num_phi])
ylim([-0.1, 0.3])

savefig('comp_norm_schemes.pdf')
show()


# plot different subtraction schemes
figure()

title('Comparison of Subtraction Schemes')

plot(sbs_true_corr / sbs_true_corr[0], lw=2)
plot(ttl_true_corr / ttl_true_corr[0], lw=2)
legend(['Shot-by-shot subtraction', 'Ensemble mean subtraction'])

xlim([0, r.num_phi])
ylim([-0.05, 0.2])

savefig('comp_subtraction_schemes.pdf')
show()
```
