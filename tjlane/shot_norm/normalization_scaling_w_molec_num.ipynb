```python
from odin import xray
from odin import structure

xray.logger.propagate = False
xray.scatter.logger.propagate = False
```

```python
# simulate gold at increasing numbers of molecules per shot

gold = structure.load_coor('fcc_sphere_5.0nm.coor')

r1 = xray.Rings.simulate(gold, 1, [2.67, 3.08], 360, 100)
r2 = xray.Rings.simulate(gold, 2, [2.67, 3.08], 360, 100)
r4 = xray.Rings.simulate(gold, 4, [2.67, 3.08], 360, 100)
r8 = xray.Rings.simulate(gold, 8, [2.67, 3.08], 360, 100)
r16 = xray.Rings.simulate(gold, 8, [2.67, 3.08], 360, 100)
```

```python
# compute correlators, both un-normalized (norm=4)
# and STD normalized (norm=3)

c1 = r1.correlate_intra(2.67, 3.08, 4)
c2 = r2.correlate_intra(2.67, 3.08, 4)
c4 = r4.correlate_intra(2.67, 3.08, 4)
c8 = r8.correlate_intra(2.67, 3.08, 4)
c16 = r16.correlate_intra(2.67, 3.08, 4)

c1s = r1.correlate_intra(2.67, 3.08, 3)
c2s = r2.correlate_intra(2.67, 3.08, 3)
c4s = r4.correlate_intra(2.67, 3.08, 3)
c8s = r8.correlate_intra(2.67, 3.08, 3)
c16s = r16.correlate_intra(2.67, 3.08, 3)


# plot just the raw correlation function w/o normalization
figure()
plot(c1.mean(0), lw=2)
plot(c2.mean(0), lw=2)
plot(c4.mean(0), lw=2)
plot(c8.mean(0), lw=2)
plot(c16.mean(0), lw=2)
show()
```

```python
# compute TJ's sqrt normalization factor
d1 = float( sqrt( product( r1.intensity_profile()[:,1]) ) )
d2 = float( sqrt( product( r2.intensity_profile()[:,1]) ) )
d4 = float( sqrt( product( r4.intensity_profile()[:,1]) ) )
d8 = float( sqrt( product( r8.intensity_profile()[:,1]) ) )
d16 = float( sqrt( product( r8.intensity_profile()[:,1]) ) )
print d1, d2, d4, d8, 16
```

```python
# now plot sqrt-normed correlators (left) side-by-side w/STD
# normed correlators (right)

figure(figsize=(15,8))

subplot(121)
title('Sqrt Norm')
plot( c1.mean(0) / d1 )
plot( c2.mean(0) / d2 )
plot( c4.mean(0) / d4 )
plot( c8.mean(0) / d8 )
plot( c16.mean(0) / d16 )

xlim([0, 360])
xlabel(r'\Delta')
ylabel('Normalized Cross Correlation')

subplot(122)
title('STD Norm')
plot( c1s.mean(0) )
plot( c2s.mean(0) )
plot( c4s.mean(0) )
plot( c8s.mean(0) )
plot( c16s.mean(0) )

xlim([0, 360])
xlabel(r'\Delta')

show()
```
